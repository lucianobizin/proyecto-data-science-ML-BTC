# Diccionario

* BTC_metals_2 => exportada como BTC_metals


- **Date**: fecha en YYYY-MM-DD (datetime64 [ns])
- **Price**: precio de cierre de BTC en el día de la fecha (USD) (float) 
- **Open**: precio de apertura de BTC en el día de la fecha (USD) (float) 
- **High**: precio más alto de BTC en el día de la fecha (USD) (float) 
- **Low**: precio más bajo de BTC en el día de la fecha (USD) (float) 
- **Vol.**: volumen de BTC transferidos (comprados+vendidos) en el día de la fecha (BTC) (float) 
- **Porcentage_diff**: diferencial porcentual del precio de BTC en la fecha [x+1] con respecto a la fecha [x] (float)  
- **Target**: 1 indica que en el día de la fecha el precio subió, y 0 que el precio bajó (float) (se puede convertir en int).
- **Price_gold**: indica el precio de cierre del oro en el día de la fecha (float).
- **Perc_diff_gold**: diferencial porcentaul del precio de la fecha [x+1] con respecto a la fecha [x] (float)
- **Price_silver**: indica el precio de cierre de la plata en el día de la fecha (float).
- **Perc_diff_silver**: diferencial porcentual del precio de la plata en la fecha [x+1] con respecto a la fecha [x] (float)
- **Price_copper**: indica el precio de cierre del cobre en el día de la fecha (float).
- **Perc_diff_copper**:  diferencial porcentaul del precio del cobre en la fecha [x+1] con respecto a la fecha [x] (float)
- **Price_aluminium**: indica el precio de cierre del aluminio en el día de la fecha (float).
- **Perc_diff_aluminium**:  diferencial porcentaul del precio del aluminio en la fecha [x+1] con respecto a la fecha [x] (float)
- **Price_platinum**: indica el precio de cierre del platinum en el día de la fecha (float).
- **Perc_diff_platinum**:  diferencial porcentaul del precio del platino en la fecha [x+1] con respecto a la fecha [x] (float)
- **Price_palladium**: indica el precio de cierre del paladio en el día de la fecha (float).
- **Perc_diff_palladium**: diferencial porcentaul del precio del paladio en la fecha [x+1] con respecto a la fecha [x] (float)
