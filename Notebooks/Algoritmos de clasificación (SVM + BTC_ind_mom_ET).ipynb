- ADX_DX: valores del índice de movimiento direccional (DX) (float) (Variable numérica continua)
- ADX_PDI: valores del Corriente Plus Movimiento Direccional (PDM) (float) (Variable numérica continua)
- ADX_PDM: valores del Máximo Actual Menos el Máximo Anterior (+DM) (float) (Variable numérica continua)
- Aroon_Oscillator: valores del Oscilador Aroon (float) (Variable numérica continua)

- MACDSIGNAL: valores de la señal de MACD (float) (Variable numérica continua)
- MACDHIST: valores del histograma del MACD (float) (Variable numérica continua)
- MOM: valores del indicador Momentum (float) (Variable numérica continua)
- ROC: valores de la tasa de cambio de una variable durante un período específico de tiempo (float) (Variable numérica continua)
- RSI: valores del índice de fuerza relativa (float) (Variable numérica continua)
- STOCHRSI_fastk: valores de la aplicación del método estocástico al indicador RSI (con variable d) (más lento) (float) (Variable numérica continua)
- ULTOSC: valores del indicador Ultimate Oscillator (float) (Variable numérica continua)
- WILLR: valores del indicador Williams %R (float) (Variable numérica continua)
- ADX_ET: valores categóricos de la aplicación de la estrategia ADX (comprar, vender, mantener posición) (object)
- ADXR_ET: valores categóricos de la aplicación de la estrategia ADXR (comprar, vender, mantener posición) (object)

Al no tratarse de un problema de regresión lineal del valor que tomaría el precio de BTC en el registro siguiente sino un problema de clasificación, la eliminación de los valores NaN no afectaría en demasía la predicción del modelo a pesar de la autocorrelación referida en el punto 2. ## Se genera un modelo PCA

```python
# Se genera la variable X (variables) e y (target)
X = BTC_ind_mom_2.drop ('Target', axis=1)
y = BTC_ind_mom_2 ['Target']
```

```python
# Se separa el data set en train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Se normalizan los datos para que PCA otorgue mejores resultados
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Se aplica PCA
pca = PCA()
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

```python
# Se analiza la varianza de cada componente
explained_variance = pca.explained_variance_ratio_
explained_variance
```

Observaciones:

```python
# Se genera la variable X (variables) e y (target)
X = BTC_ind_mom_2.drop ('Target', axis=1)
y = BTC_ind_mom_2 ['Target']
```

```python
# Se separa el data set en train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Se normalizan los datos para que PCA otorgue mejores resultados
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Se genera un nuevo PCA con 22 componentes (las variables que recogen el 97 % de la varianza de los datos)
pca = PCA(n_components=22)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

```python
# Se definen los parámetros de búsqueda
param_grid= {'kernel': ['rbf'],
             'C': [0.1, 1, 10, 100, 1000],
             'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
             'random_state':np.arange(1,5000,50),
             'class_weight': ['balanced'] }

from sklearn.model_selection import RandomizedSearchCV
grid_SVM = RandomizedSearchCV(SVC(), param_grid, refit=True, verbose=3, n_iter=100, cv=5)
```

```python
grid_SVM.fit(X_train, y_train)
```

```python
# print best parameter after tuning
print(grid_SVM.best_params_)
  
# print how our model looks after hyper-parameter tuning
print(grid_SVM.best_estimator_)
```

```python
# Se crea el modelo SVM 
modelo = SVC(C = 100, kernel = "rbf", gamma=0.01, random_state=4101, class_weight="balanced")
```

```python
# Se entrena al modelo
modelo.fit(X_train, y_train)
```

```python
# Se busca conocer las predicciones del modelo
y_train_pred = modelo.predict(X_train)
y_test_pred = modelo.predict(X_test)
```

```python
# Se calcula el accuracy en los datos de train
train_accuracy = accuracy_score(y_train, y_train_pred)
print('% de aciertos sobre el set de evaluación para train:', train_accuracy)

# Se calcula el accuracy en los datos de test
test_accuracy = accuracy_score(y_test, y_test_pred)
print('% de aciertos sobre el set de evaluación para test:', test_accuracy)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix

# Predicciones
predictions = modelo.predict(X_test)

#Matriz de Confusión
print(confusion_matrix(y_test, predictions))

#Ploteamos la Matriz
plot_confusion_matrix(modelo, X_test, y_test)
plt.show()
```

Observaciones:
- El modelo genera muchos casos de falsos negativos y positivos.

```python
# Accuracy
print(f"Accuracy of the classifier is: {accuracy_score(y_test, predictions)}")

# Precision
from sklearn.metrics import precision_score
print(f"Precision Score of the classifier is: {precision_score(y_test, predictions)}")

# Recall
from sklearn.metrics import recall_score
print(f"Recall Score of the classifier is: {recall_score(y_test, predictions)}")

# Calculo del F1 score
from sklearn.metrics import f1_score
print(f"F1 Score of the classifier is: {f1_score(y_test, predictions)}")
```

Observaciones:
- El mejor modelo predice en un 63,6% de los casos si el precio de BTC sube o baja.
