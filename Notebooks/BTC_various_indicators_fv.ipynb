Paso 2 -> se agrega a BTC_various indicators_fv las variables con mayor correspondencia (>70) con las variables Price, Percentage_diff y Target en cada base de datos. Variables y grado de correspondencia (ver archivos de análisis univariado, bivariado y multivariado para cada data set):

- BTC_metals_fv_df: Price_gold (76), Price_silver (77), Price_copper (88), Price_aluminium (76).
- BTC_ind_trend_ET_fv_df: SMA7 (1), SMA 15 (1), EMA_15 (1), EMA_15 (1), Bollinger bands (1) (¿dejar 0 y 1 con otro nombre?).
- BTC_ind_mom_ET_fv_df: ADX_PDM (91).
- BTC_multiple_indicators_fv_df: ATR (91), Linear Regression (1), Linear intercept (98), STDDEV (84), TSF (1), Trend (1). 
- BTC_pattern_indicators_fv_df: N/A.
- BTC_US_fv_df: Price_SP500 (86), Price_Nasdaq (89), Price_DJ (82).

```python
BTC_various_indicators_fv = BTC_metals_fv_df [["Date", "Price", "Open", "High", "Low", "Vol.", 
                                               "Percentage_diff","Target", "Price_gold", "Price_copper","Price_aluminium",
                                               "Trend", "Residuals"]].copy()
```

Se dejan solo las variables seleccionadas de cada base de datos para luego hacer un merge.

```python
BTC_ind_trend_ET_fv_df_2 = BTC_ind_trend_ET_fv_df [['Date', 'EMA_7d', 'HT_TRENDLINE', 'BBANDS_upper_band',
       'BBANDS_middle_band', 'BBANDS_lower_band', 'PSAR', 'EMA_7d_ET',
       'HTTL_ET', 'PSAR_ET']].copy()
```

```python
BTC_ind_mom_ET_fv_df_2 = BTC_ind_mom_ET_fv_df [['Date', 'ADX_PDM']].copy()
```

```python
BTC_multiple_indicators_fv_df_2 = BTC_multiple_indicators_fv_df [['Date', 'ATR', 'Linearregression', 'Linearintercept', 'STDDEV']].copy()
```

```python
BTC_US_fv_df_2 = BTC_US_fv_df [['Date', 'Price_SP500', 'Price_NASDAQ', 'Price_DJ']].copy()
```

Se agregan las variables restantes utilizando merge a partir de la variable ["Date"]

```python
BTC_various_indicators_fv = BTC_various_indicators_fv.merge (BTC_ind_trend_ET_fv_df_2, on="Date", how="inner")
```

```python
BTC_various_indicators_fv = BTC_various_indicators_fv.merge (BTC_ind_mom_ET_fv_df_2, on="Date", how="inner")
```

```python
BTC_various_indicators_fv = BTC_various_indicators_fv.merge (BTC_multiple_indicators_fv_df_2, on="Date", how="inner")
```

```python
BTC_various_indicators_fv = BTC_various_indicators_fv.merge (BTC_US_fv_df_2, on="Date", how="inner")
```

```python
BTC_various_indicators_fv.head()
```

```python
BTC_various_indicators_fv.columns
```

Paso 3 -> se modifican las variables dummies por variables numéricas discretas y se eliminan las variables categóricas.

```python
Variables = ['EMA_7d_ET','HTTL_ET', 'PSAR_ET']

for var in Variables:
    dummies = pd.get_dummies(BTC_various_indicators_fv [var])
    dummies = dummies.rename(columns={"Comprar":"Comprar_"+var+"", "Mantener_posicion":"Mantener_posicion_"+var+"", "Vender":"Vender_"+var+""})
    BTC_various_indicators_fv = pd.concat([BTC_various_indicators_fv, dummies], axis = 1)
```

```python
BTC_various_indicators_fv.head()
```

```python
BTC_various_indicators_fv = BTC_various_indicators_fv.drop(['EMA_7d_ET','HTTL_ET', 'PSAR_ET'], axis=1)
```
