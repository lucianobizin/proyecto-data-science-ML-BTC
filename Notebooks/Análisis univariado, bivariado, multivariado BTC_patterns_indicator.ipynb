Análisis de variable ["Date"] y ["Precio"] segúin los elementos de una serie de tiempo

- Componentes de una serie de tiempo: Xt = Tt  (tendencia) + St (estacionariedad) + Ct (ciclicidad) + Rt (residualidad)

En este caso concreto:
- Ciclidad = son variaciones cíclicas en el precio de BTC sin período de tiempo definido (suelen ser más largos que los de estacionariedad). 
- Tendencia = incremento o decrecimiento del precio de BTC a largo plazo (no tiene porqué ser lineal y puede variar). 
- Estacionariedad = son patrones repetitivos de precio con un periodo definido (se relacionan con un fenómeno -> ej. estaciones del año, etc.).
- Residualidad/residuos = son fluctuaciones del precio de BTC que son de carácter irregular e impredecibles.

```python
# Se aplica el filtro Hodrick-Prescott para separar la tendencia y el componente ciclico del precio de BTC
BTC_cycle, BTC_trend = sm.tsa.filters.hpfilter(BTC_pattern_df['Close'])
```

```python
# Se crea un dataframe con BTC_ciclo y se lo indexa según las fechas del dataframe original
BTC_cycle_df = pd.DataFrame(data=BTC_cycle.values, index=BTC_pattern_df ["Date"], columns=["BTC_cycle"])
```

```python
BTC_cycle_df.head()
```

```python
# Se grafica la ciclicidad
sns.set(rc={'figure.figsize':(20, 5)})
sns.lineplot(x=BTC_cycle_df.index, y=BTC_cycle_df ["BTC_cycle"]);
```

```python
# Se grafica el ciclo (ampliado para los últimos 365 días)
BTC_cycle_df_2 = BTC_cycle_df.tail(365)
sns.set(rc={'figure.figsize':(20, 5)})
sns.lineplot(x=BTC_cycle_df_2.index, y=BTC_cycle_df_2 ["BTC_cycle"]);
```

Observaciones:
- Se puede observar que la ciclidad del precio de BTC es variante a lo largo del tiempo.

```python
#Otra forma de descomponer la serie de tiempo utilizando statsmodels (freq=1 día)
descomposicion = sm.tsa.seasonal_decompose(BTC_cycle_df ['BTC_cycle'], model='additive')
```

```python
# Graficando la descomposición de la serie de tiempo: precio, tendencia, estacionariedad y residualidad (así se completa el análisis de los 4 elementos)
fig = descomposicion.plot()
fig.set_size_inches((15, 8))
fig.tight_layout()
plt.show()
```

Observaciones:

- La tendencia y el precio se mueven en paralelo (como era de esperarse). 
- Al menos se puede decir que hay dos grandes tendencias (una alcista desde el 2016 a los últimos meses de 2021, y una bajista desde ahí en adelante).
- Analizando a simple vista el gráfico de precio de BTC en el tiempo (arriba) se puede percibir que no es una serie estacionaria (su varianza, media y estructura de autoregresión se modifican con el tiempo). Esto significa que la serie necesita ser suavizada y tratada para el momento de ser modelada (los modelos de ML suelen trabajar con el supuesto de estacionariedad). Al aplicarse sm.tsa.seasonal_decompose ahora se percibe estacionariedad (Seasonal) dado que ya fue realizado dicho proceso.
- La estacionariedad se dibujará debajo para poder entender lo que se ha graficado como un solo bloque azul.
- La residualidad aumenta con el movimiento más brusco del precio (2018, 2021 y 2022 fueron los años de mayor crecimiento/decrecimiento de precio).

Se demuestra que su varianza no fue siempre la misma => serie no estacionaria

```python
# Se obtienen los datos de varianza para diferentes períodos de tiempo. 
# Se entiende que si la varianza no se modificó a lo largo del tiempo la cantidad de datos analizados
# no debería influir sustancialmente en los resultados obtenidos.

BTC_var = []
x=0
while x < len (BTC_pattern_df ["Close"]):
    var = BTC_pattern_df ["Close"] [0:x].var()
    BTC_var.append (var)
    x+=1
    
BTC_var_df = pd.DataFrame (data=BTC_var, index=BTC_pattern_df.index, columns=["Variance"])
```

```python
sns.lineplot(x=BTC_var_df.index, y=BTC_var_df ["Variance"]);
```

Observación:
- La varianza se ha modificado a lo largo del tiempo.

Se analiza la variabilidad de la media

```python
# Se obtienen los datos de media para diferentes períodos de tiempo. 
# Se entiende que si la media no se modificó a lo largo del tiempo, la cantidad de datos analizados
# no debería influir sustancialmente en los resultados obtenidos.
BTC_mean = []
x=0
while x < len (BTC_pattern_df ["Close"]):
    var = BTC_pattern_df ["Close"] [0:x].mean()
    BTC_mean.append (var)
    x+=1
    
BTC_mean_df = pd.DataFrame (data=BTC_mean, index=BTC_pattern_df.index, columns=["Mean"])
```

```python
sns.lineplot(x=BTC_mean_df.index, y=BTC_mean_df ["Mean"]);
```

Observación:
- La media no fue constante => serie no estacionaria. 

Se estudia la estructura de autocorrelación de la serie de tiempo

```python
# Lag_plot se utiliza principalmente para encontrar patrones en series de tiempo.
# URL: https://pandas.pydata.org/docs/reference/api/pandas.plotting.lag_plot.html
# Se analizará la autocorrelación entre los valores de precio a lo largo del tiempo 
# para diferentes diferencias (1 día de atraso, 7 días, etc).
variables = range(1,200,6)

columns = 6

fig, axes = plt.subplots((len(variables)+2) // (columns), columns, figsize=(15,30))

for current_idx, variable in enumerate(variables):
    i = current_idx // columns
    j = current_idx % columns
    lag_plot(BTC_pattern_df ["Close"], lag=variable, c="orange", ax=axes[i][j])
    axes[i][j].set_title(variable)
    axes[i][j].set_xlabel("")
    
plt.tight_layout()
```

Observaciones:
- Se nota que existe correlación entre los valores de precio del día x con los valores de precio del día x-n (n=días de atraso). A medida que "n" es mayor, la correlación disminuye considerablemente. Para los n= 181, 187, 193 y 199, la correlación es muy pobre.

```python
# tsaplots.plot_acf => plotea la función de autocorrelación.
# URL: https://www.statsmodels.org/dev/generated/statsmodels.graphics.tsaplots.plot_acf.html
# Los lags son graficados en el eje x, y la correlación en el eje y.
# Se analiza la correlación con distintos lags (1 a 200).

variables = range(1,200,6)

columns = 6

fig, axes = plt.subplots((len(variables)+2) // (columns), columns, figsize=(15,30))

for current_idx, variable in enumerate(variables):
    i = current_idx // columns
    j = current_idx % columns
    fig = tsaplots.plot_acf (BTC_pattern_df ["Close"], lags=variable, color="g", ax=axes[i][j])
    axes[i][j].set_title(variable)
    axes[i][j].set_xlabel("")
    
plt.tight_layout()
```

Observaciones:
- La autocorrelación va disminuyendo con el tiempo, e inversamente, el grado de error del alpha (intervalo de confianza) también va aumentando. De una correlación de 1 se termina en una correlación de 0.7 a los 200 días. De un alpha menor a 0.2 se termina en un alpha superior a 0.5. 

```python
# Se grafica la autocorrelación parcial del precio según un intervalo de confianza de 0.05.
# URl: https://www.statsmodels.org/dev/generated/statsmodels.graphics.tsaplots.plot_pacf.html
sm.graphics.tsa.plot_pacf(BTC_pattern_df ["Close"], method='ywadjusted', alpha=0.05)
plt.show()
```

Observación:
- La serie de tiempo no posee una estructura de autoregresión constante => no es estacionaria.
- La autocorrelación máxima se da con un día de atraso (el valor previo tiene una alta incidencia en el precio siguiente).
- Sin embargo, la autocorrelación parcial nos permite detectar lags que también están por fuera de la zona de confianza (5, 8, 9, 10, 11, 13, etc.). Estos números pueden ayudar a masterizar el modelo de ML (hiperparámetros).

```python
# Otra forma de probar que es una serie no estacionaria -> Dickey-Fuller test.
# URl: https://www.statsmodels.org/dev/generated/statsmodels.tsa.stattools.adfuller.html
from statsmodels.tsa.stattools import adfuller
sm.tsa.stattools.adfuller(BTC_pattern_df ["Close"])
```

Observaciones:
- La p-value => 0.43 > 0.05 => serie no estacionaria.
